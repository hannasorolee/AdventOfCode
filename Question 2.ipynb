### Question 1. How many passwords are valid according to their policies?

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("question2_rawinput.csv", header=None, names=['PWS'])

df
```

```python

def get_range(password):
    return password.split(' ')[0]

def get_instances(password):
    return password.split(' ')[1]

def get_password(password):
    return password.split(' ')[2]

df['range'] = df['PWS'].apply(lambda x: f"{get_range(x)}") #lambda x to grab cell content and f"() for formatting

df['instances'] = df['PWS'].apply(lambda x: f"{get_instances(x)}") #lambda x to grab cell content and f"() for formatting
df['instances']=df['instances'].str[0:1]

df['password'] = df['PWS'].apply(lambda x: f"{get_password(x)}")


df.head()
```

```python
def get_minrange(range):
    return range.split('-')[0]

def get_maxrange(range):
    return range.split('-')[1]

df['min'] = df['range'].apply(lambda x: f"{get_minrange(x)}") #lambda x to grab cell content and f"() for formatting

df['max'] = df['range'].apply(lambda x: f"{get_maxrange(x)}") #lambda x to grab cell content and f"() for formatting


df

```

```python
df['min'] = df['min'].astype('int32')
df['max'] = df['max'].astype('int32')
df['password'] = df['password'].astype('str')
df.info()
```

```python
def countingmethod(passwordcol, instancecol):
    return passwordcol.count(instancecol)

df['count_instance'] = df.apply(lambda x: countingmethod(x['password'], x['instances']), axis=1)
df
```

```python
def countvalids(count_instanceCol, maxCol, minCol):
    if count_instanceCol <= maxCol and count_instanceCol >= minCol:
        valid = 1
    else:
        valid = 0
    return valid


df['valid'] = df.apply(lambda x: countvalids(x['count_instance'], x['max'], x['min']), axis=1)
df
```

```python
sum(df['valid'])
```

### Question 2: How many passwords are valid according to the new interpretation of the policies?

```python
df['password'].str[2]
```

```python
df['min_index']=df['min']-1
df['max_index']=df['max']-1
df
```

```python
def password_minvalid(passwordCol, maxCol, minCol, instancesCol):

    if passwordCol[minCol]  == instancesCol:
        valid = 1
    else:
        valid = 0
    
    return valid


df['min_valid'] = df.apply(lambda x: password_minvalid(x['password'], x['max_index'], x['min_index'], x['instances']), axis=1)


def password_maxvalid(passwordCol, maxCol, minCol, instancesCol):

    if passwordCol[maxCol]  == instancesCol:
        valid = 1
    else:
        valid = 0
    
    return valid


df['max_valid'] = df.apply(lambda x: password_maxvalid(x['password'], x['max_index'], x['min_index'], x['instances']), axis=1)
df.head(50)
```

```python
df['final']=df['max_valid']+df['min_valid']

sum(x == 1 for x in df['final'])
```
