### Question 1: they need you to find the two entries that sum to 2020 and then multiply those two numbers together.

```python
import pandas as pd
import numpy as np
import itertools
```

```python
df = pd.read_csv("question1_rawinput.csv", header=None, names=['Numbers'])

df.head()

df
```

```python
K = 2020

def findDouble(df, K):
    res = []
    for var in itertools.combinations(df['Numbers'], 2):
        if var[0] + var[1] == K:
            res.append((var[0], var[1]))
    return res

print(findDouble(df ,K))
```

### Question 2: What is the product of the three entries that sum to 2020?

```python
K = 2020

def findTriple(df, K):
    res = []
    for var in itertools.combinations(df['Numbers'], 3):
        if var[0] + var[1] + var[2] == K:
            res.append((var[0], var[1], var[2]))
    return res

print(findTriple(df ,K))
```
