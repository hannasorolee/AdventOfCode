### Question 1

```python
import pandas as pd 

with open("day_4_input.txt") as f:
    input = f.readlines()
input
```

```python
input = [line.strip() for line in input]
    
input
```

```python

matches = ["byr", "iyr", "eyr", "hgt", "hcl", "ecl", "pid"]

def cons(response):
    questions = 0
    if all(x in response for x in matches): #https://stackoverflow.com/questions/3389574/check-if-multiple-strings-exist-in-another-string
        questions = 1
    else:
        questions = 0
            
    return questions
```

```python
new_sum = 0
currentResponse = ''

for line in input:
    if line != '':
        currentResponse += line
    
    else:
        new_sum +=cons(currentResponse)
        currentResponse = ''

new_sum += cons(currentResponse)

print(new_sum)
```
