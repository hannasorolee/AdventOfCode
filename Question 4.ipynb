```python
with open("day_4_input.txt") as f:
    input = f.read()


groups = input.split('\n\n')
groups
```

```python

matches = ["byr", "iyr", "eyr", "hgt", "hcl", "ecl", "pid"]

def cons(response):
    questions = 0
    if all(x in response for x in matches): #https://stackoverflow.com/questions/3389574/check-if-multiple-strings-exist-in-another-string
        questions = True
    else:
        questions = False
            
    return questions
```

### Question 2:

```python
import re 
def valid_hgt(value):
    if len(value) < 4:
        return False
    if value[-2:] not in ('cm', 'in'):
        return False
    if not value[:-2].isdigit():
        return False
    if value[-2:] == 'cm':
        if not (150<= int(value[:-2]) <= 193):
            return False
    if value[-2:] == 'in':
        if not (59<= int(value[:-2]) <= 76):
            return False
    return True
```

```python
def valid_hcl(value):
    if len(value) != 7:
        return False
    if value[0] != '#':
        return False
    if not bool(re.search("[a-f0-9]{6}", value[1:])):
        return False
    return True


valid_eye_colors = ["amb","blu","brn","gry","grn","hzl","oth"]
```

```python
def validate_fields(field, value):
    
    #byr, iyr, eyr
    if field == 'byr':
        if not len(value) == 4:
            return False
        value = int(value)
        if not (1920 <= value <= 2002):
            return False
    if field == 'iyr':
        if not len(value) == 4:
            return False
        value = int(value)
        if not (2010 <= value <= 2020):
            return False
    if field == 'eyr':
        if not len(value) == 4:
            return False
        value = int(value)
        if not (2020 <= value<= 2030):
            return False
    
    #hgt
    if field == 'hgt':
        if valid_hgt(value) is not True:
            return False
    
    #hcl
    if field == 'hcl':
        if valid_hcl(value) is not True:
            return False
    
    #ecl
    if field == 'ecl':
        if value not in valid_eye_colors:
            return False
    
    #pid
    if field == 'pid':
        if len(value) != 9:
            return False
        if not value.isdigit():
            return False
    return True
        
```

```python
p1 = 0
unvalidated_passports=0
validate_passports = 0
number_passports = 0


for group in groups: #per passport
    number_passports +=1
    valid = True
    
    if cons(group): #Question 1 - validity 
        p1+=1
        
        g = group.split('\n') #splitting every space until it hits last word of string

        for p in g: #per string
            y = p.split(' ') #split their passport by space to get all passport metrics as each strings
        
            for z in y: #per metrics
                n = z.split()

                for b in n:
                    kv = b.split(":")
                
                    if validate_fields(kv[0], kv[1]) is False:
                        valid = False
                    
                    #print(kv[0],kv[1])
                    #print(validate_fields(kv[0], kv[1]))
    else:
        valid = False
        
    if valid == True:
        print("valid", g)
        validate_passports +=1
        print(validate_passports)
    else:
        unvalidated_passports +=1
        #print("unvalid", g)

print("Question 1", p1)
print("Question 2", validate_passports)
print("Question 2", unvalidated_passports)
print("number of passports", number_passports)
```
